## **Part (a): Dimensionless equation of motion**

$$V(x) = 
\begin{cases}
    \infty & x<0 \\
    \frac{1}{2}m\omega^2 - V_0 & 0 \le x \le x_b \\
    0 & x > x_b
\end{cases}$$

$$\tau = \omega t \implies \frac{d}{dt} = \omega \frac{d}{d\tau} \tag{a.1}$$
$$\xi = \frac{x}{x_b} 
\implies x = x_b \xi \tag{a.2}$$
$$\implies \frac{dx}{dt} = \dot{x} = x_b \frac{d\xi}{dt} = x_b\omega\frac{d\xi}{d\tau}$$
$$\implies \dot{x} = x_b \omega \xi' \tag{a.3}$$
$$\ddot{x} = x_b\omega^2 \xi'' \tag{a.4}$$

**Equation of motion:**
$$F = m\ddot{x} = -\frac{dV}{dx} \\
\implies m\omega^2x_b\xi'' = -\frac{dV}{dx} \tag{from a.4}$$
$$\implies \xi'' = -\frac{1}{m\omega^2x_b}\frac{dV}{dx}$$

Now,
$$\frac{dV}{dx} = 
\begin{cases}
    m\omega^2x & 0 \le x \le x_b \\
    0 & x>x_b
\end{cases} \\[10 pt]
\implies -\frac{1}{m\omega^2x_b}\frac{dV}{dx} = \xi'' = 
\begin{cases}
    -\xi & 0 \le \xi \le 1 \\
    0 & \xi > 1
\end{cases}$$

**Continuity of V:**
$$V(\xi) = 
\begin{cases}
    \infty & \xi<0 \\
    \frac{1}{2}m\omega^2x_b^2\xi^2 - V_0 & 0 \le \xi \le 1 \\
    0 & \xi>1
\end{cases} \\[10 pt]
= 
\begin{cases}
    \infty & \xi < 0 \\
    V_0(\xi^2-1) & 0\le\xi\le1 \\
    0 & \xi > 1
\end{cases}$$

At $\xi = 1$:

Right Hand limit (RHL):
$$V(1^+) = 0$$

Left Hand limit (LHL):
$$V(1^-) = V_0(1^2 - 1) = 0$$

Since RHL = LHL, $V$ is continuous at $\xi = 1$

$$\varepsilon = \frac{1}{2}\left(\frac{1}{2}\right)^2 - \frac{1}{2} = -\frac{3}{8} = -0.375$$

At turning points, $\xi' = 0$. As can be seen from the given data, for $\varepsilon = -0.375$, one of the turning points is at $\xi = 0.5$. The other will be at $\xi = -0.5$. But since we are only working on the positive region, hence $\xi = 0.5$ and $\xi = 0$ (hard wall at x = 0) are the turning points.

## **Part (c): Bound states and shooting method**

```python
# RK4
def rk4(f1, x, y, h):
    k1 = h * f1(x, y)
    k2 = h * f1(x + h / 2, y + k1 / 2)
    k3 = h * f1(x + h / 2, y + k2 / 2)
    k4 = h * f1(x + h, y + k3)
    return y + (k1 + 2 * k2 + 2 * k3 + k4) / 6

# Dimensionless potential
def dim_V(y):
    return 1/2*y[0]**2 - 1/3*y[0]**3

# Total energy
def tot_energy(y):
    return 1/2*y[1]**2 + dim_V(y)

# Caller for RK4 method
def caller1(fn, y_ini, N, xs, h):
    y = y_ini
    ys = np.zeros((N, len(y_ini)), dtype=np.float64)
    Es = np.zeros(N)
    
    for i in range(N):
        x = xs[i]
        ys[i, :] = y
        Es[i] = tot_energy(y)
        y = rk4(fn, x, y, h)
     
    return ys, Es

# d/dt[e e']
def f(t, y):
    return np.asarray([y[1], -y[0]+y[0]**2])
```
