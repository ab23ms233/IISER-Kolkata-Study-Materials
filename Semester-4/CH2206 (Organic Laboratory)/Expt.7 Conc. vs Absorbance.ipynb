```python
import matplotlib.pyplot as plt
```

```python
x = [i for i in range(5, 26, 5)]
y = list(map(lambda x: 0.0294*x, x))
print(x)
print(y)
```

```python
plt.plot(x, y, ls = "solid", color = "green", marker = "o")
plt.xlabel("Concentartion(in mM)")
plt.ylabel("Absorbance")
plt.title("Alpha cyanostilbene")
plt.show()
```
