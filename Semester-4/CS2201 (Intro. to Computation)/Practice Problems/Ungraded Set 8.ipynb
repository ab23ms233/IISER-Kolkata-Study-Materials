```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.integrate as spi
import math
```

Q1

```python
lower, upper = 1,2
intervals = 50
no_of_pts = 100

def f(x):
    return 1/x

step = (upper-lower)/(intervals-1)

xval = np.linspace(lower, upper, num=no_of_pts)
yval = np.apply_along_axis(f, 0, xval)
```

RECTANGLE METHOD

```python
integration_rect = 0
count = 0
x = lower
val_list = []

while x <= upper:
    val = f(x)
    val_list.append(val)
    
    area = val*step
    integration_rect += area
    
    rect_x = np.array([x for i in range(2)])
    
    if count == 0:
        rect_y = np.array([0, val])
    else:
        rect_y = np.array([0, val_list[count-1]])
        
    plt.plot(rect_x, rect_y, color = "green", ls = "dashed")
    plt.plot([x, x+step], [val, val], color="green", ls="dashed")
    
    x += step
    count += 1

print(integration_rect)

plt.plot(xval, yval, ls = "solid")
plt.show()
```

TRAPEZOIDAL RULE

```python
integration_trap = 0
x = lower+step
count = 0

plt.plot([lower, lower], [0, f(lower)], color="green", ls = "solid")
plt.plot([lower, x], [f(lower), f(x)], color="green")

while x<upper:
    val = f(x)
    integration_trap += step*val
    
    rect_x = np.array([x for i in range(2)])
    rect_y = np.array([0, val])
        
    plt.plot(rect_x, rect_y, color = "green", ls = "solid")
    
    if x+step <=upper: 
        plt.plot([x, x+step], [f(x), f(x+step)], color="green", ls="solid")
    
    x += step
    count+=1
    
integration_trap += step*(f(lower) + f(upper))/2
plt.plot(xval, yval, color="blue", ls = "solid")
print(integration_trap)
plt.show()
```

SIMPSON'S RULE

```python
integration_simp = 0
x = lower + step
count = 1

while x < upper:
    val = f(x)
    
    if count%2 == 1:
        integration_simp += 4*step*val/3
    else:
        integration_simp += 2*step*val/3
    
    x += step
    count += 1
    
integration_simp += step*(f(lower)+f(upper))/3
print(integration_simp)
```

```python
actual = math.log(2)

print("Trapezoidal Rule: ", np.trapz(yval, xval))
print("Simpson's Rule: ", spi.simpson(yval, xval))
print("Actual Value: ", actual)
print()
print("Error in rectangle method: ", abs(integration_rect - actual))
```
