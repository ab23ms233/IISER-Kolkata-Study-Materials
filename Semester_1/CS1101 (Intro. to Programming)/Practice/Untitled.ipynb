```python
import math
```

```python
x = 1
num_of_terms = 1
sign = 1
denominator = 1
numerator = x
computed_pi = 0
```

```python
# round(abs(4*computed_pi-math.pi), 3)

while round(abs(4*computed_pi-math.pi), 3) != 0:
    term = sign*numerator/denominator
    computed_pi = computed_pi + term
    
    denominator += 2
    numerator = numerator*x**2
    sign *= -1
    num_of_terms += 1

print(num_of_terms)
print("Exact value of pi:", math.pi)
print("Computed value of pi:", 4*computed_pi)
print(abs(4*computed_pi-math.pi))
```
