Q1

```python
import matplotlib.pyplot as plt

months = eval(input("Enter months in list format: "))
rainfall = eval(input("Enter rainfall in list format: "))

plt.plot(months, rainfall)
plt.xlabel("Months")
plt.ylabel("Rainfall")
plt.title("Rainfall v/s Months")
plt.show()
```

Q2

```python
temperature = eval(input("Enter temperature in list format: "))

plt.plot(months, temperature, color="blue")
plt.plot(months, rainfall, color = "green")
plt.xlabel("Months")
plt.ylabel("Temperature/Rainfall")
plt.title("Months v/s Temperature/Rainfall")
plt.legend(["Temperature", "Rainfall"], loc="upper right")
plt.show()
```

```python
fig, (axm, axt) = plt.subplots(2)

axm.plot(months, rainfall, "gs-")
axm.set_xlabel("Months")
axm.set_ylabel("Rainfall")
axm.set_title("Months v/s Rainfall")

axt.plot(months, temperature, "bo-")
axt.set_xlabel("Months")
axt.set_ylabel("Temperature")
axt.set_title("Months v/s Temperature")

fig.tight_layout()
plt.show()
```

```python
import random
import math

def toss():
    num = random.random()
    
    if num < 0.5:
        return "H"
    else:
        return "T"

num_of_trials = int(math.pow(2, 13))
H_count = 0
P_H = []
trials = [i for i in range(1, num_of_trials+1)]

for i in range(num_of_trials):
    result = toss()
    
    if result == "H":
        H_count += 1
    
    P_H.append(H_count/num_of_trials)

plt.plot(trials, P_H)
plt.xlabel("Number of Trials")
plt.ylabel("Probability of Head")
plt.title("Coin Toss")

plt.show()
```
