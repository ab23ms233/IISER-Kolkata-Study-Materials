```python
import matplotlib.pyplot as plt
```

Q1

```python
'''months = eval(input("Enter months in list: "))
rainfall = eval(input("Enter rainfall in list: "))'''

months = [1,2,3,4,5]
rainfall = [34,56,23,15,42]
temperature = [63,21,34,32,41]

plt.plot(months, rainfall)
plt.xlabel("Months")
plt.ylabel("Rainfall")
plt.title("Graph")

plt.show()
```

Q2

```python
plt.plot(months, rainfall, color="green", ls="dotted")
plt.plot(months, temperature, color="red", ls="-")
plt.legend(["Rainfall", "Temperature"])
plt.show()
```

Q4a

```python
fig, (axr, axt) = plt.subplots(1, 2)

axr.plot(months, rainfall)
axr.set_xlabel("Months")
axr.set_ylabel("Rainfall")
axr.set_title("Graph 1")

axt.plot(months, temperature)
axt.set_xlabel("Months")
axt.set_ylabel("Temperature")
axt.set_title("Graph 2")

fig.tight_layout()
plt.show()
```

Q4b

```python
fig, (axr, axt) = plt.subplots(2, 1)

axr.plot(months, rainfall)
axr.set_xlabel("Months")
axr.set_ylabel("Rainfall")
axr.set_title("Graph 1")

axt.plot(months, temperature)
axt.set_xlabel("Months")
axt.set_ylabel("Temperature")
axt.set_title("Graph 2")

fig.tight_layout()
plt.show()
```

Q5

```python
plt.bar(months, rainfall)
plt.show()
```

Q6

```python
plt.bar(months, rainfall, 0.4, label = "rainfall")
plt.bar(months, temperature, 0.4, label="temperature")
plt.legend(["Rainfall", "Temperature"])

plt.show()
```

Q7

```python
total_rainfall = sum(rainfall)
percent_rainfall = []

for i in range(len(rainfall)):
    percent = rainfall[i]/total_rainfall*100
    percent_rainfall.append(percent)
    
plt.pie(percent_rainfall, labels=months, autopct='%1.2f%%')
plt.show()
```
