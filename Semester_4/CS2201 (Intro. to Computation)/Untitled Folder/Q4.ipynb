Preparing the database

```python
#Database of sentences
database = ["what a nice weather", "it is a good weather", "i see a gigantic building", "i see a huge building", "what a gorgeous flower", "it was a fast delivery", "what a beautiful flower", "it was a quick delivery"]
word = "quick"

#Words in each sentence
sent_words = []

#All unique words
all_words = []

#Duplicate words
duplicate_words = []

for sentence in database:
    semantics = sentence.split()
    sent_words.append(semantics)
    
    for word in semantics:
        if word not in all_words:
            all_words.append(word)
        else:
            if word not in duplicate_words:
                duplicate_words.append(word)
            

print(sent_words)
print()
print(all_words)
print()
print(duplicate_words)
```

```python
#Removing duplicate words from all words
for i in range(len(duplicate_words)):
    word = duplicate_words[i]
    
    if word in all_words:
        all_words.remove(word)

print(all_words)
```

```python
#word = input("Enter a word: ")
sent_word = sent_words[:]

#Checking semantics of input word
def semantics(word):
    word_semantics = []
    
    for words in sent_word:
        for i in range(len(words)):
            if words[i] == word:
                word_semantics = words[:i] + words[i+1:]
                return word_semantics
    
    if len(word_semantics) == 0:
        print("Word is not present in database")

known_semantic = semantics(word)

#Removing input word from all_words
all_words.remove(word)
print(semantics(word))
```

```python
common_list = []

#Finding the number of common words in between semantics
for i in range(len(all_words)):
    semantic = semantics(all_words[i])
    common = 0
    
    for i in range(min(len(semantic), len(known_semantic))):
        if semantic[i] == known_semantic[i]:
            common += 1
    common_list.append(common)

print(common_list)
```

```python
max_index = common_list.index(max(common_list))
print("Semantically similar word is", all_words[max_index])
```
