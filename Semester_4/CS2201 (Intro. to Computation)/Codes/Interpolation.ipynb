```python
import matplotlib.pyplot as plt
from scipy import interpolate
import numpy as np
```

```python
start, stop, step = 0, 10, 0.5

def f(x):
    return x**2

x_known = np.arange(start, stop, step)
y_known = f(x_known)
```

```python
print(f"Known x-values: {x_known}")
print(f"Known y-values: {y_known}")
```

INTERPOLATION 1D

```python
model = interpolate.interp1d(x_known, y_known)
start_new, stop_new, step_new = 0, 9, 0.2

x_new = np.arange(start_new, stop_new, step_new)
y_new = model(x_new)
```

```python
print(f"New x-values: {x_new}")
print(f"Predicted y-values: {y_new}")
```

```python
plt.scatter(x_known, y_known, color="red", marker="o", label="Known")
plt.plot(x_new, y_new, ls = "dashed", color="green", label="Predicted")
plt.title("Interpolation")
plt.legend()
plt.show()
```

NEWTON FORWARD INTERPOLATION

```python
def forward_diff_table(y_known):
    n = len(y_known)
    diff_table = [y_known.copy()]
    
    for i in range(1, n):
        new_diff = [diff_table[i-1][j+1] - diff_table[i-1][j] for j in range(n - i)]
        diff_table.append(new_diff)
        
    return diff_table

def newton_forward_interpolation(x_known, y_known, x_to_interpolate):
    h = x_known[1] - x_known[0]
    u = (x_to_interpolate - x_known[0]) / h

    diff_table = forward_diff_table(y_known)

    result = y_known[0]
    u_term = 1
    
    for i in range(1, len(x_known)):
        u_term *= (u - i + 1) / i
        result += u_term * diff_table[i][0]

    return result

# Example usage
x_vals = [1.0, 1.1, 1.2, 1.3]
y_vals = [1.0, 1.21, 1.44, 1.69]

x_interp = 1.15
y_interp = newton_forward_interpolation(x_vals, y_vals, x_interp)

print(f"Interpolated value at x = {x_interp} is approximately {y_interp:.4f}")
```

LAGRANGE INTERPOLATION

```python
def lagrange_interpolation(x_values, y_values, x_to_interpolate):
    n = len(x_values)
    result = 0.0

    for i in range(n):
        term = y_values[i]
        for j in range(n):
            if i != j:
                term *= (x_to_interpolate - x_values[j]) / (x_values[i] - x_values[j])
        result += term

    return result

# Example data
x_vals = [1.0, 1.3, 1.6, 1.9]
y_vals = [1.0, 1.69, 2.56, 3.61]  # y = x^2

x_interp = 1.5
y_interp = lagrange_interpolation(x_vals, y_vals, x_interp)

print(f"Interpolated value at x = {x_interp} is approximately {y_interp:.4f}")
```
