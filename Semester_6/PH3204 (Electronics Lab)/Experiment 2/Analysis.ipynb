```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

Input Characteristics

```python
# Loading data
vce2 = pd.read_csv("Data/inputvce2.csv")
vce3 = pd.read_csv("Data/inputvce3.csv")
vce4 = pd.read_csv("Data/inputvce4.csv")

# Plotting
plt.figure()
plt.plot(vce2.iloc[:,0], vce2.iloc[:,1], label='$V_{CE}$ = 2 V')
plt.plot(vce3.iloc[:,0], vce3.iloc[:,1], label='$V_{CE}$ = 3 V')
plt.plot(vce4.iloc[:,0], vce4.iloc[:,1], label='$V_{CE}$ = 4 V')

plt.xlabel('$V_{BE}$ (V)')
plt.ylabel('$I_B$ (µA)')
plt.title("Input characteristics of CE configuration")
plt.legend()
plt.grid()

plt.show()
```

Output Characteristics

```python
# Loading data
ib10 = pd.read_csv("Data/outputib10.csv")
ib20 = pd.read_csv("Data/outputib20.csv")
ib30 = pd.read_csv("Data/outputib30.csv")
ib40 = pd.read_csv("Data/outputib40.csv")

# Plotting
plt.figure()
plt.plot(ib10.iloc[:,0], ib10.iloc[:,1], label='$I_{B}$ = 10 µA')
plt.plot(ib20.iloc[:,0], ib20.iloc[:,1], label='$I_{B}$ = 20 µA')
plt.plot(ib30.iloc[:,0], ib30.iloc[:,1], label='$I_{B}$ = 30 µA')
plt.plot(ib40.iloc[:,0], ib40.iloc[:,1], label='$I_{B}$ = 40 µA')

plt.xlabel('$V_{CE}$ (V)')
plt.ylabel('$I_C$ (mA)')
plt.title("Output characteristics of CE configuration")
plt.legend()
plt.grid()

plt.show()
```

Fitting Data for calculating $I_{C}$ (mA) and $\beta$ at $V_{CE}$ = 5V

```python
x_data = [[2.88, 5.93], [1.19, 4.3, 8.15], [2.12, 6.55], [1.17, 4.55]]
y_data = [[1.972, 2.1], [3.7, 3.8, 3.9], [5.1, 5.4], [7, 7.5]]
ms, cs = [], []
ibs = [10, 20, 30, 40]

def straight_line(x, m, c):
    return m*x + c

for i in range(len(x_data)):
    popt, pcov = curve_fit(straight_line, x_data[i], y_data[i])
    m, c = popt

    ms.append(m)
    cs.append(c)
    ic = straight_line(5, m, c)
    ib = ibs[i]

    print(f"Ic for Ib = {ib} uA: {ic:0.3f} ")
    print(f"Beta = {ic/ib:0.3f}")
    print()
```
